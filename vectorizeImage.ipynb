# Image Vectorize

This will convert all the images in the database to vectors using the model created for image searching

```python
import tensorflow as tf
from tensorflow.keras.models import load_model
```

```python
vectorizer = load_model('model.h5')
```

```python
import os
```

```python
from PIL import Image
```

```python
images = os.listdir('static/imageDB')
```

```python
images
```

```python
imagesVectorList = [tf.convert_to_tensor(Image.open('static/imageDB/'+img).resize((256, 256)).convert('RGB')) for img in images]
```

```python
imagesVector = tf.stack(imagesVectorList)
```

```python
representation = vectorizer.predict(imagesVector)
```

```python
import numpy as np
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarityVector = cosine_similarity(representation[0].reshape(1, -1), representation).tolist()[0]
similarityVector
```

```python
imagesSimilarity = list(zip(images, similarityVector))
```

```python
rankedImages = sorted(imagesSimilarity, key=lambda x: x[1], reverse=True)
```

```python
[img[0] for img in rankedImages]
```
